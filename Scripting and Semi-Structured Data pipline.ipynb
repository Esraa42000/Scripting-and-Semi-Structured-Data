```python

import pandas as pd
import numpy as np
import httpagentparser
```

```python
def converted(filepath):
  df = pd.read_json(filepath, lines=True)
  o=[]
  b=[]
  for index in df.a:
    o.append(httpagentparser.simple_detect(index)[0])
    b.append(httpagentparser.simple_detect(index)[1])
  cleaned=pd.DataFrame(columns=[])
  cleaned['browser']=o
  cleaned['os']=b
  cleaned['from_url'] = df['r'].str.split(pat=r'/', expand=True)[2]
  cleaned.from_url.fillna('direct',inplace=True)
  cleaned['to_url'] = df['u'].str.split(pat=r'/', expand=True)[2]
  cleaned['city']=df['cy']
  cleaned['city']=cleaned["city"].fillna("not found")
  df['ll']=df['ll'].fillna("not found")
  df['ll']=df['ll'].astype('str')
  cleaned['Longitude'] = df['ll'].str.strip('[]').str.split(',',expand=True)[0]
  cleaned['Latitude'] = df['ll'].str.strip('[]').str.split(',',expand=True)[0]
  cleaned.Longitude.replace(np.nan,"not found",inplace=True)
  cleaned.Latitude.replace(np.nan,"not found",inplace=True)
  clean=df['tz'].fillna("not found")
  clean[clean==""]="not found"
  cleaned['Time zone']=clean
  cleaned['time_in']=pd.to_datetime(df["t"], unit='ms')
  cleaned['time_out']=pd.to_datetime(df["hc"], unit='ms')
  return cleaned
```

```python
data_csv=pd.DataFrame(columns=[])
file=df = "/content/drive/MyDrive/pythondata/usa.gov_click_data.json"
data_csv=converted(file)
data_csv.to_csv('website_info.csv',index=False)
display(data_csv)
```
